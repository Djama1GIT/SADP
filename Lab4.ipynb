```python
# Определяем класс Node для узлов бинарного дерева поиска
class Node:
    def __init__(self, value):
        self.value = value    # Значение узла
        self.left_child = None    # Левый дочерний узел
        self.right_child = None    # Правый дочерний узел

# Определяем класс BinarySearchTree для бинарного дерева поиска
class BinarySearchTree:
    def __init__(self):
        self.root = None    # Корневой узел

    # Метод insert() для вставки нового узла в дерево
    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(value, self.root)

    # Рекурсивный метод _insert() для вставки нового узла в дерево
    def _insert(self, value, current_node):
        if value < current_node.value:
            if current_node.left_child is None:
                current_node.left_child = Node(value)
            else:
                self._insert(value, current_node.left_child)
        elif value > current_node.value:
            if current_node.right_child is None:
                current_node.right_child = Node(value)
            else:
                self._insert(value, current_node.right_child)
        else:
            print("Value already exists in tree.")

    # Метод search() для поиска узла в дереве по значению
    def search(self, value):
        if self.root is None:
            return False
        else:
            return self._search(value, self.root)

    # Рекурсивный метод _search() для поиска узла в дереве по значению
    def _search(self, value, current_node):
        if current_node is None:
            return False
        elif current_node.value == value:
            return True
        elif value < current_node.value:
            return self._search(value, current_node.left_child)
        else:
            return self._search(value, current_node.right_child)

    # Метод delete() для удаления узла из дерева по значению
    def delete(self, value):
        if self.root is None:
            print("Tree is empty.")
        else:
            self.root = self._delete(value, self.root)

    # Рекурсивный метод _delete() для удаления узла из дерева по значению
    def _delete(self, value, current_node):
        if current_node is None:
            return None
        elif value < current_node.value:
            current_node.left_child = self._delete(value, current_node.left_child)
            return current_node
        elif value > current_node.value:
            current_node.right_child = self._delete(value, current_node.right_child)
            return current_node
        else:
            if current_node.left_child is None and current_node.right_child is None:
                return None
            elif current_node.left_child is None:
                return current_node.right_child
            elif current_node.right_child is None:
                return current_node.left_child
            else:
                min_value = self._find_min_value(current_node.right_child)
                current_node.value = min_value
                current_node.right_child = self._delete(min_value, current_node.right_child)
                return current_node

    # Рекурсивный метод _find_min_value() для поиска минимального значения в дереве
    def _find_min_value(self, current_node):
        while current_node.left_child is not None:
            current_node = current_node.left_child
        return current_node.value


# Создаем экземпляр класса BinarySearchTree
tree = BinarySearchTree()

# Вставляем каждый элемент из списка values в дерево
for value in values:
    tree.insert(value)

# Выводим дерево на экран
print(tree)
start_time = timeit.default_timer()
tree.search(234)
end_time = timeit.default_timer()
print("Время работы алгоритма составило:", binttime := (end_time - start_time), "секунд")
# Вставляем элемент 7 в дерево и проверяем, найден ли он в дереве
tree.insert(7)
print(f"Элемент 7{' не' if not tree.search(7) else ''} найден в дереве")

# Удаляем элемент 7 из дерева и проверяем, найден ли он в дереве
tree.delete(7)
print(f"Элемент 7{' не' if not tree.search(7) else ''} найден в дереве")
```

```python
# Рехэширование с помощью псевдослучайных чисел

class HashTable:
    def __init__(self):
        # Инициализация пустой хеш-таблицы размером 10
        self.size = 20
        self.table = [[] for _ in range(self.size)]

    def hash_function(self, key):
        # Хеш-функция, вычисляющая хеш-значение ключа
        return key % self.size

    def put(self, key, value):
        # Добавление нового элемента в хеш-таблицу
        hash_value = self.hash_function(key)
        bucket = self.table[hash_value]
        for i in range(len(bucket)):
            if bucket[i][0] == key:
                # Если элемент с таким ключом уже существует, обновляем его значение
                bucket[i] = (key, value)
                return
        # Если элемент с таким ключом не существует, добавляем его в конец корзины
        bucket.append((key, value))

    def get(self, key):
        # Получение значения элемента по ключу из хеш-таблицы
        hash_value = self.hash_function(key)
        bucket = self.table[hash_value]
        for i in range(len(bucket)):
            if bucket[i][0] == key:
                # Если элемент с заданным ключом найден, возвращаем его значение
                return bucket[i][1]
        # Если элемент с заданным ключом не найден, возвращаем None
        return None

    def delete(self, key):
        # Удаление элемента по ключу из хеш-таблицы
        hash_value = self.hash_function(key)
        bucket = self.table[hash_value]
        for i in range(len(bucket)):
            if bucket[i][0] == key:
                # Если элемент с заданным ключом найден, удаляем его из корзины
                del bucket[i]
                return

    def __str__(self):
        # Возвращает строковое представление хеш-таблицы в виде списка списков
        return str(self.table)


h = HashTable()

h.put(1, "apple")
h.put(2, "banana")
h.put(3, "cherry")

print(h.get(1))  # output: apple
print(h.get(2))  # output: banana
print(h.get(3))  # output: cherry

h.delete(2)
print()
print(h.get(1))  # output: apple
print(h.get(2))  # output: None
print(h.get(3))  # output: cherry


def test_hash_table():
    h = HashTable()

    # Тестирование методов put() и get()
    for i in range(100):
        key = random.randint(1, 100)
        value = ''.join(random.choices(string.ascii_lowercase, k=5))
        h.put(key, value)
        assert h.get(key) == value

    # Тестирование метода delete()
    for i in range(50):
        key = random.randint(1, 100)
        h.delete(key)
        assert h.get(key) is None

    # Тестирование перехеширования
    for i in range(100):
        key = random.randint(101, 200)
        value = ''.join(random.choices(string.ascii_lowercase, k=5))
        h.put(key, value)
        assert h.get(key) == value

    # Тестирование замены существующих ключей
    for i in range(50):
        key = random.randint(1, 100)
        value = ''.join(random.choices(string.ascii_lowercase, k=5))
        h.put(key, value)
        assert h.get(key) == value


test_hash_table()
```

```python
# Метод цепочек
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class HashTable:
    def __init__(self, capacity=100):
        self.capacity = capacity  # Количество ячеек в хэш-таблице
        self.size = 0  # Текущее количество элементов в хэш-таблице
        self.buckets = [None] * self.capacity  # Массив с элементами хэш-таблицы

    def hash(self, key):
        hash_sum = 0
        for idx, char in enumerate(key):
            # Вычисляем хэш-сумму ключа
            hash_sum += (idx + len(key)) ** ord(char)
            # Ограничиваем хэш-сумму размером хэш-таблицы
            hash_sum = hash_sum % self.capacity
        return hash_sum

    def insert(self, key, value):
        index = self.hash(key)  # Вычисляем индекс элемента в хэш-таблице
        node = self.buckets[index]
        while node:
            if node.key == key:  # Если элемент с таким ключом уже существует, заменяем его значение
                node.value = value
                return
            node = node.next
        # Создаем новый элемент и добавляем его в начало списка
        new_node = Node(key, value)
        new_node.next = self.buckets[index]
        self.buckets[index] = new_node
        self.size += 1

    def search(self, key):
        index = self.hash(key)  # Вычисляем индекс элемента в хэш-таблице
        node = self.buckets[index]
        while node:
            if node.key == key:  # Если элемент с таким ключом найден, возвращаем его значение
                return node.value
            node = node.next
        # Если элемент не найден, возвращаем None
        return None

    def delete(self, key):
        index = self.hash(key)  # Вычисляем индекс элемента в хэш-таблице
        node = self.buckets[index]
        prev = None
        while node:
            if node.key == key:  # Если элемент с таким ключом найден, удаляем его
                if prev:
                    prev.next = node.next
                else:
                    self.buckets[index] = node.next
                self.size -= 1
                return node.value
            prev = node
            node = node.next
        # Если элемент не найден, возвращаем None
        return None


# Создаем новую хэш-таблицу
table = HashTable()

# Добавляем элементы в хэш-таблицу
table.insert('apple', 1)
table.insert('banana', 2)
table.insert('cherry', 3)

# Ищем элемент в хэш-таблице
print(table.search('apple'))  # 1

# Удаляем элемент из хэш-таблицы
table.delete('banana')

# Проверяем, что элемент удален
print(table.search('banana'))  # None

import random

# Создаем новую хэш-таблицу
table = HashTable()

# Заполняем хэш-таблицу случайными значениями
for i in range(1000):
    key = str(random.randint(0, 100000))
    value = random.randint(0, 100000)
    table.insert(key, value)

# Проверяем, что все элементы добавлены корректно
for i in range(1000):
    key = str(random.randint(0, 100000))
    value = random.randint(0, 100000)
    table.insert(key, value)
    assert table.search(key) == value

# Удаляем случайные элементы из хэш-таблицы
for i in range(500):
    key = str(random.randint(0, 100000))
    table.delete(key)
    assert table.search(key) is None

# Проверяем, что все оставшиеся элементы найдены корректно
for i in range(1000):
    key = str(random.randint(0, 100000))
    value = random.randint(0, 100000)
    table.insert(key, value)
    assert table.search(key) == value
```

Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.

```python
def solve_n_queens(n):
    # вложенная функция для рекурсивного перебора всех возможных расстановок ферзей
    def dfs(queens, xy_dif, xy_sum):
        # p - количество уже расставленных ферзей
        p = len(queens)
        # если расставлены все ферзи, то добавляем текущую расстановку в список решений
        if p == n:
            result.append(queens)
            return None
        # перебираем все возможные позиции для следующего ферзя
        for q in range(n):
            # проверяем, что эта позиция не конфликтует с уже расставленными ферзями
            if q not in queens and p - q not in xy_dif and p + q not in xy_sum:
                # если позиция допустима, то продолжаем рекурсивный перебор с новой расстановкой
                dfs(queens + [q], xy_dif + [p - q], xy_sum + [p + q])

    # список для хранения решений
    result = []
    # вызываем рекурсивную функцию для поиска всех возможных расстановок ферзей
    dfs([], [], [])
    return result


def print_board(queens):
    board = []
    for i in range(len(queens)):
        row = ['□  '] * len(queens)
        row[queens[i]] = '●  '
        board.append(''.join(row))
    # выводим расстановку ферзей на доску
    print('\n'.join(board))


# пример использования
solutions = solve_n_queens(8)
print_board(solutions[0])

# проверка нахождения хотя бы одного решения для 12 ферзей
solutions = solve_n_queens(12)
assert len(solutions) > 0

# проверка корректности решения для 12 ферзей
for solution in solutions:
    for i in range(len(solution)):
        for j in range(i + 1, len(solution)):
            assert solution[i] != solution[j]
            assert abs(solution[i] - solution[j]) != abs(i - j)
```
