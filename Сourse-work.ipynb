# Задание
Написать программу, которая будет считывать данные из CSV файла, содержащего
информацию о продажах товаров в магазине. 

Данные в файле содержатся в следующем
формате:

| Номер заказа | Дата заказа | Название товара | Категория товара | Количество продаж | Цена

Необходимо:
1. Рассчитать общую выручку магазина.
2. Найти товар, который был продан наибольшее количество раз.
3. Найти товар, который принес наибольшую выручку.
4. Составить отчет, содержащий информацию об общей выручке магазина, количестве проданных единиц каждого товар и доле каждого товара в общей выручке.

Также необходимо учитывать возможные ошибки ввода-вывода и обрабатывать их в
соответствии с требованиями.

# Выполнение

```python
def str_to_darray(string, separator=" "):
    separators = [" ", " ", "\t", "\r", "\n"]
    darray = []
    tempstring = ""
    start = 0
    end = - 1
    for i in range(len(string)):
        if string[i] not in separators:
            start = i
            break
    for i in range(len(string) - 1, -1, -1):
        if string[i] not in separators:
            end = i
            break
    string = string[start:end + 1]
    for i in string:
        if i == separator:
            if tempstring != "":
                darray += [tempstring]
            tempstring = ""
        else:
            tempstring += i
    if tempstring:
        darray += [tempstring]
    return [darray]
```

```python
def quick_sort(array, index=0, reverse=False):
    if len(array) > 1:
        pivot = array.pop()
        lst1, equal_lst, lst2 = [], [pivot], []
        for item in array:
            if (item[index] == pivot[index]):
                equal_lst += [item]
            elif (item[index] < pivot[index]) if reverse else (item[index] > pivot[index]):
                lst1 += [item]
            else:
                lst2 += [item]
        return (quick_sort(lst2, index, reverse) + equal_lst + quick_sort(lst1, index, reverse))
    else:
        return array
```

```python
def print_table(_table: [list[str], list[str], ...]):
    row_sizes = [0] * len(_table[0])
    for i in _table:
        for j, v in enumerate(i):
            v = str(v)
            row_sizes[j] = row_sizes[j] if row_sizes[j] > len(v) else len(v)
    _print_table = [[] for _ in range(len(_table))]
    for i, vi in enumerate(_table):
        for j, vj in enumerate(vi):
            vj = str(vj)
            _print_table[i] += [f"{' ' * (row_sizes[j] // 2 + row_sizes[j] % 2 - len(vj) // 2 - len(vj) % 2)}{vj}"
                                f"{' ' * (row_sizes[j] // 2 - len(vj) // 2)} |"]
    space_count = (sum([len(cell) for cell in _print_table[0]]) + len(_print_table[0]))
    print('_' * space_count)
    [(print('|', end=""), print(*row), print('|' + '_' * (space_count - 2) + '|')) for row in _print_table]
```

```python
with open('table.csv', 'r', encoding='utf-8') as file:
    table = []
    for i in file:
        # table += [i.strip().split(';')]
        table += str_to_darray(i, ';')
```

# Отчет

```python
all_money = 0
most_popular = (-1, -1)
best_profit = (-1, -1)
headers = ["Общая выручка магазина",
           "Товар, который был продан наибольшее количество раз",
           "Товар, который принес наибольшую выручку"]

for i in range(1, len(table)):
    all_money += int(table[i][-1])
    most_popular = most_popular \
        if (m := (int(table[i][4]), -i)) < most_popular else m
    best_profit = best_profit \
        if (t := (int(table[i][4]) * int(table[i][5]), i)) < best_profit else t
report = [[*table[0][2:6], "Выручка", "Доля"]]
for i in range(1, len(table)):
    report += [[*table[i][2:6], revenue := int(table[i][4]) * int(table[i][5]),
                revenue / all_money]]
```

```python
def print_sorted_report():
    _report = quick_sort(report[1:], index=5, reverse=True)
    for i in _report:
        i[-1] = f"{i[-1] * 100}"[:5] + "%"
    print_table([report[0], *_report])
```

```python
print_table([headers,
             [all_money,
              table[-most_popular[1]][2],
              table[best_profit[1]][2]]])
print_sorted_report()
```
